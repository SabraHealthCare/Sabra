```python
import pandas as pd
import os
import math
from docx.enum.table import WD_TABLE_ALIGNMENT
import numpy as np
import pandas
from statistics import mean
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams

import datetime
from docx import Document
from docx.shared import Inches
from docx.shared import Cm
from docx.shared import Pt
from docx.shared import RGBColor
from docx.enum.text import WD_ALIGN_PARAGRAPH
from os import walk

mypath="C://Users//sli//Downloads//Survey"
changename="No"
os.chdir(mypath)
rcParams.update({'figure.autolayout': True})
def insert_table(df_table,word_size,title,lengthlist=[]):
        
        global table_number
        # get table data -------------nametable
        x=df_table.shape[0]+1
        y=df_table.shape[1]
        table = document.add_table(x,y)
        table.alignment = WD_TABLE_ALIGNMENT.CENTER
        table.style = 'TableGrid'

        for i in range(x):
            for j in range(y):
                if i==0:
                    table.rows[i].cells[j].text=str(df_table.columns[j])
                else:
                    table.rows[i].cells[j].text=str(df_table.iloc[i-1,j])
                if j<len(lengthlist):
                    table.rows[i].cells[j].width = Cm(lengthlist[j])

                
        for row in table.rows:
            for cell in row.cells:
                paragraphs = cell.paragraphs
                for paragraph in paragraphs:
                    for run in paragraph.runs:
                        font = run.font
                        font.size= Pt(word_size)

        
        p = document.add_paragraph()
        p.alignment = 1
        run=p.add_run(title)
        table_number+=1
        font=run.font
        font.size = Pt(12)

        

filename_list=[]
for (dirpath, dirnames, filenames) in walk(mypath):
    filename_list.extend(filenames)
    break
```

```python
for filename in filename_list:
    season=["Summer","Spring","Fall","Winter","Total"]
    doc_season=list(filter(lambda x:x.lower() in filename.lower(),season))[0]
    year=["2019", "2020","2021","2022", "2023", "2024", "2025", "2026", "2027", "2028", "2029"]
    doc_year=list(filter(lambda x:x in filename,year))[0]
    term=["Midterm","Final",""]
    doc_term=str(list(filter(lambda x:x.lower() in filename.lower(),term))[0])
    type=["Course","Instructor"]
    course=["BSN","MSN","PharmD","Clinical","Pharmacy","MSCR","Site","GE"]
    doc_course=list(filter(lambda x:x.lower() in filename.lower(),course))
    
    data = pd.read_excel(filename)
    data['InstructorName']=data['InstructorName'].apply(lambda x:x.replace("Dr. ",""))
    data['InstructorName']=data['InstructorName'].apply(lambda x:x.replace("Dr.",""))
    data['InstructorName']=data['InstructorName'].apply(lambda x:x.replace("M.S",""))
    data['InstructorName']=data['InstructorName'].apply(lambda x:x.replace("MS.",""))
    for i in range(data.shape[0]):
        # split “,”
        if "," in str(data["InstructorName"].iloc[i]):
            InstructorName_list=data["InstructorName"].iloc[i].split(",")
            data.loc[i,"InstructorName"]=InstructorName_list[0].strip()
            
            for j in range(1,len(InstructorName_list)):
                data=data.append(data.iloc[i,:],ignore_index=True)
                data.loc[len(data)-1,"InstructorName"]=InstructorName_list[j].strip()
    file=data
    individuals = data["InstructorName"].unique()
    for person in individuals:
        document = Document()


        for survery_type in range(0,2):
            data=file
            data=data[data["InstructorName"]==person]
            table_number=1   
            # Process data
            file_name=" ".join([doc_year,doc_season,doc_term])+" "+" ".join(doc_course)+" "+type[survery_type]+" Evaluation"+"_"+person
            document.add_heading(file_name, 0)
            if type[survery_type]=="Course":
                question_number=['Q2','Q3','Q4','Q5','Q6','Q7','Q8','Q9']
            elif type[survery_type]=="Instructor":
                question_number=['Q10','Q11','Q12','Q13','Q14','Q15']


            #改question名
            for i in range(data.shape[1]):
                if "Question" in data.columns[i]:
                    data.rename(columns={data.columns[i]:("Q"+data.columns[i][-2:]).replace(" ", "")},inplace=True)


            # 去除 NA
            data.dropna(subset=question_number,inplace=True)
            data.reset_index(drop=True,inplace=True)
            if len(data)==0:
                break

            #去除0
            for i in range(data.shape[0]):
                for j in range(data.shape[1]):
                    if data.iloc[i,j]==0:
                        data.iloc[i,j]=None

            

      # Participation      
            p=document.add_heading(level=1)
            p.add_run('Survey participation ').font.size=Pt(22) 
            document.add_paragraph(style='Intense Quote')


# table 1: Course and insturctor Name
            if survery_type==0:
                realname=data["CourseTitle"].unique()
                labelname=[type[survery_type]+str(i) for i in np.arange(1,len(realname)+1)]
                nametable=pd.DataFrame(labelname,columns=[" "])
                nametable["CourseTitle"]=pd.DataFrame(realname)
                coursetable=data.groupby(["CourseTitle","InstructorName"]).size().reset_index().rename(columns={0:"# of Response"})
                mapping=pd.merge(nametable,coursetable,how="outer")
                mapping.rename(columns={0:type[survery_type]})
                table_title="Table "+str(table_number)+": Course and Instructor Name"
                insert_table(mapping,10,table_title,[2,8,5,2])
            elif survery_type==1:
                realname=data["InstructorName"].unique()
                labelname=[type[survery_type]+str(i) for i in np.arange(1,len(realname)+1)]
                nametable=pd.DataFrame(labelname,columns=[" "])
                nametable["InstructorName"]=pd.DataFrame(realname)
                coursetable=data.groupby(["InstructorName","CourseTitle"]).size().reset_index().rename(columns={0:"# of Response"})
                mapping=pd.merge(nametable,coursetable,how="outer")
                mapping=mapping.drop([" "],axis=1)
                table_title="Table "+str(table_number)+": Course and Instructor Name"
                insert_table(mapping,13,table_title,[5,5,5])
                data["CourseTitle"]=data["InstructorName"]
           
 
           
        
        #隐藏 course 名
            if changename=="Yes":
                #realname=data["CourseTitle"].unique()
                #labelname=[type[survery_type]+str(i) for i in np.arange(1,len(realname)+1)]
                for i in range(data.shape[0]):
                    for j in range(len(realname)):
                        if realname[j]==data["CourseTitle"][i]:
                            data["CourseTitle"][i]=labelname[j]
                            break
            elif survery_type==0:
                data["CourseTitle"]=data["CourseCode"].apply(lambda x:x.split()[0])
                data["CourseTitle"]=data["CourseCode"].apply(lambda x:x.split('_')[0])
            elif survery_type==1:
                data["CourseTitle"]=data["InstructorName"].apply(lambda x: x.replace("Prof.","").split()[0])

            data["Coursename"]=data["CourseTitle"]

           
            data["Average"]=data[question_number].apply(mean,axis=1)
            respond_number=data["CourseTitle"].dropna().count()
            course_number=data["CourseTitle"].unique().shape[0]


            #document.add_picture('count response.png', width=Inches(4))
            p = document.add_paragraph()
            run=p.add_run(str(course_number)+" "+str(type[survery_type].lower())+'s are evaluated in this survey.  Total '+str(respond_number)+' survey responds. ')
            font=run.font
            font.size=Pt(13)
            p.alignment = WD_ALIGN_PARAGRAPH.LEFT
            last_paragraph = document.paragraphs[-1] 
            last_paragraph.alignment = WD_ALIGN_PARAGRAPH.CENTER
      



    # Table 2: Average Score
            n=data.groupby("CourseTitle")[question_number].mean().round(2)
            bardata=n.apply(np.mean,axis=1).sort_values()

            p=document.add_heading(level=1)
            p.add_run('Average Score for '+type[survery_type].lower()).font.size=Pt(22) 
            document.add_paragraph(style='Intense Quote')
           
            last_paragraph = document.paragraphs[-1] 
            last_paragraph.alignment = WD_ALIGN_PARAGRAPH.CENTER
            p = document.add_paragraph()
            font = run.font
            font.size = Pt(16)

            
            x=question_number.copy()
            x.append("Average")
            m=data.groupby("Coursename")[x].mean().round(2)
         #m.reset_index(drop=False,inplace=True).rename(columns={0:type[survery_type]})
            m.reset_index(drop=False,inplace=True)
            m.rename(columns={"Coursename":type[survery_type]},inplace=True)
            table_title='Table '+str(table_number)+': Average score of each question for '+type[survery_type]
            insert_table(m,10,table_title)

            document.add_page_break()


            m=data.groupby("Coursename")[x].mean().round(2)
            m = m.T

            plt.figure(figsize=(5,3)) 
            #ax = m.plot.bar(rot=0, subplots=True)
            #plt.title("Average score for "+type[survery_type],size=15)
            ax = m.plot.bar(rot=0, subplots=True)
            #plt.xticks(rotation=0,size=15)
            #plt.ylim(ymin=1,ymax=5)
  
            #plt.yticks(size=15)
            plt.xlabel("Survey Question Number")
            plt.savefig('Averages_score.png', dpi=800)

            document.add_picture('Averages_score.png', width=Inches(4))
            last_paragraph = document.paragraphs[-1] 
            last_paragraph.alignment = WD_ALIGN_PARAGRAPH.CENTER
            p = document.add_paragraph()
            font = run.font
            font.size = Pt(16)



    
    
    
    # Score distribution
            Scorecount=pd.DataFrame(columns=["5","4","3","2","1"])
            Scorecount.set_index=data["Coursename"].unique()
            score=6
            for k in ["5","4","3","2","1",]:
                score=score-1
                for i in  data["Coursename"].unique():
                    total=data[data["Coursename"]==i].shape[0]*len(question_number)

                    Scorecount.loc[i,k]="{:0.0%}".format(round(data[data["Coursename"]==i][question_number].apply(lambda x:x==score).sum().sum()/total,2))

            p=document.add_heading(level=1)
            p.add_run('Score distribution').font.size=Pt(22) 
            document.add_paragraph(style='Intense Quote')

            p = document.add_paragraph()
            p.alignment = WD_ALIGN_PARAGRAPH.CENTER
            run=p.add_run("1: Low score    5: High score")
            font = run.font
            font.size= Pt(14) 

            Scorecount.reset_index(drop=False,inplace=True)
            Scorecount.rename(columns={"index":type[survery_type]},inplace=True)
            table_title='Table '+str(table_number)+': Score distribution 1~5 for '+type[survery_type]
            insert_table(Scorecount,13,table_title)

            
            Scorecount=pd.DataFrame(columns=["5","4","3","2","1"])
            #plot
            Scorecount.set_index=data["Coursename"].unique()
            score=6
            for k in ["5","4","3","2","1",]:
                score=score-1
                for i in  data["Coursename"].unique():
                    total=data[data["Coursename"]==i].shape[0]*len(question_number)

                    Scorecount.loc[i,k]=round(data[data["Coursename"]==i][question_number].apply(lambda x:x==score).sum().sum()/total,2)

            Scorecount=Scorecount.T
            plt.figure(figsize=(5,3)) 
            ax = Scorecount.plot.bar(rot=0, subplots=True)
            #plt.title("Score percentage for "+type[survery_type],size=15)
            #plt.xticks(rotation=0,size=15)
            #plt.ylim(ymin=Scorecount.min()[0]-0.1,ymax=1)

            #plt.yticks(size=15)
            plt.xlabel("Score 1 ~ 5")
            plt.savefig('percent.png', dpi=800)

            document.add_picture('percent.png', width=Inches(4))
            last_paragraph = document.paragraphs[-1] 
            last_paragraph.alignment = WD_ALIGN_PARAGRAPH.CENTER
            p = document.add_paragraph()
            font = run.font
            font.size = Pt(16)

    # Low score table
            low=data.copy()
            low[question_number]=(data[question_number]<3)*1
            lowscore=low.groupby("CourseTitle")[question_number].sum()
            count=pd.DataFrame(data["CourseTitle"].value_counts())
            count.rename(columns={"CourseTitle":"Total #"})
            lowscore.join(count)

            p=document.add_heading(level=1)
            p.add_run('Low Score Counts').font.size=Pt(22) 
            document.add_paragraph(style='Intense Quote')

            p = document.add_paragraph()
            p.alignment = WD_ALIGN_PARAGRAPH.CENTER
            run=p.add_run("Low score definition:  Score of 1 or 2")
            font = run.font
            font.size= Pt(14)

            lowscore.reset_index(drop=False,inplace=True)
            lowscore.rename(columns={"CourseTitle":type[survery_type]},inplace=True)
            table_title='Table '+str(table_number)+': Low score count of each question for '+type[survery_type]
            insert_table(lowscore,13,table_title)


    # Boxplot 

            coursename=data["Coursename"].unique()
            length=len(coursename)
            if length>1:
                plt.figure(figsize=(15,5*math.ceil(length/3)))   #(18,27)
            else: 
                plt.figure(figsize=(8,5))

            for i in range(0,length):
                if length>1:
                    plt.subplot(math.ceil(length/2),2,i+1)

                coursedata=data.loc[data["Coursename"]==coursename[i]][question_number]
                coursedata.boxplot(grid=False)
                plt.yticks(np.arange(1, 5.2, step=1))
                plt.xlabel(coursename[i],fontsize=13)
                plt.tight_layout(pad=3.0)

            plt.savefig('boxplot_course.png', dpi=500)
            plt.show()


            p=document.add_heading(level=1)
            p.add_run('Boxplot of '+type[survery_type]).font.size=Pt(22) 
            document.add_paragraph(style='Intense Quote')
            document.add_picture('boxplot_course.png', width=Inches(6.5))   
            last_paragraph = document.paragraphs[-1] 
            last_paragraph.alignment = WD_ALIGN_PARAGRAPH.CENTER
            p = document.add_paragraph()
            document.add_page_break()



    # Questiontable
        Questiontable = pd.read_excel(filename, sheet_name='QuestionMapper')
        p=document.add_heading(level=1)
        p.add_run('Survey Question').font.size=Pt(22) 
        insert_table(Questiontable[0:15],10,"",[3,16])


    # Commment table
        comment=data[["InstructorName","Q16"]].dropna(how="any")
        comment.columns=["InstructorName","Comment"]
        p=document.add_heading(level=1)
        p.add_run('Comment').font.size=Pt(22) 
        table_title='Table '+str(table_number)+': Comments'
        insert_table(comment,10,table_title,[3,16])

        document.save(file_name+'.docx')
```
